```python
from sklearn import preprocessing

labels = ['mario', 'rui', 'dias', 'pires']

encoder = preprocessing.LabelEncoder()
encoder.fit(labels)

for i, item in enumerate(encoder.classes_):
    print(item, '=>', i)
```

```python
labels2 = ['mario', 'rui', 'dias', 'pires', 'rui', 'dias', 'pires', 'rui', 'dias', 'pires']
labels2_encoded = encoder.transform(labels2)

print(labels2_encoded)
```

```python
import pandas as pd

from IPython.display import display

data = pd.read_csv('adult.data', header=None, index_col=False, names=['age', 'workclass', 'fnlwgt', 'education', 
                                                                      'education-num', 'marital-status', 'occupation', 
                                                                      'relationship', 'race', 'gender', 'capital-gain', 
                                                                      'capital-loss', 'hours-per-week', 'native-country', 
                                                                      'income'])
```

```python
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]
display(data)
```

```python
print(list(data.columns))
data_dummies = pd.get_dummies(data)

print(list(data_dummies.columns))
```

```python
features = data_dummies.loc[data_dummies.index[:], 'age':'occupation_ Transport-moving']
```

```python
X = features.values
y = data_dummies['income_ >50K'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
logreg = LogisticRegression()

logreg.fit(X_train, y_train)

print(logreg.score(X_test, y_test))
```
